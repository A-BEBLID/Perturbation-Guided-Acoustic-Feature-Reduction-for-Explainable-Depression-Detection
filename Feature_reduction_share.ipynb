```python
#Keep _edaic _ Never touch 

import os
import time
import shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (classification_report, roc_auc_score, confusion_matrix, 
                             precision_score, recall_score, f1_score, ConfusionMatrixDisplay, roc_curve)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from feature_importance_keras import get_feature_importance

# === Setup ===
os.makedirs("importance_outputs", exist_ok=True)
sns.set(style="whitegrid")
epochs = 30
strategies = ['3m', 'permutation', 'median', 'mean', 'max', 'zero','hybrid']
threshold = 0.0
classification_reports = {}
results = []
strategy_importances = {}

# === Load Data ===
data = pd.read_csv('Data_edaic/04_sec.csv')
X = data.iloc[:, :-2]
y = data.iloc[:, -1]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.1, random_state=42)

# === Model Builder ===
def build_model(input_dim):
    model = Sequential([
        Dense(128, input_dim=input_dim, activation='relu'),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# === Baseline Model ===
print("Training baseline model...")
model = build_model(X_train.shape[1])
history_base = model.fit(X_train, y_train, epochs=epochs, batch_size=32, validation_split=0.2, verbose=1)

y_pred = (model.predict(X_test) > 0.5).astype(int)
y_proba = model.predict(X_test)
initial_report = classification_report(y_test, y_pred, digits=3, output_dict=True)
classification_reports['baseline'] = initial_report

initial_auc = roc_auc_score(y_test, y_proba)
initial_acc = model.evaluate(X_test, y_test, verbose=0)[1]
initial_precision = precision_score(y_test, y_pred)
initial_recall = recall_score(y_test, y_pred)
initial_f1 = f1_score(y_test, y_pred)

# === Baseline Evaluation Report ===
print("\n\U0001F4CA Baseline Classification Report")
print(classification_report(y_test, y_pred, digits=3))

cm_base = confusion_matrix(y_test, y_pred)
disp_base = ConfusionMatrixDisplay(cm_base)
disp_base.plot(cmap="Blues", values_format=".0f")
plt.title("Baseline Confusion Matrix")
plt.tight_layout()
plt.savefig("importance_outputs/confusion_matrix_baseline.png", dpi=300)
plt.show()

plt.figure()
plt.plot(history_base.history['accuracy'], label='Train Accuracy')
plt.plot(history_base.history['val_accuracy'], label='Validation Accuracy')
plt.plot(history_base.history['loss'], label='Train Loss')
plt.plot(history_base.history['val_loss'], label='Validation Loss')
plt.title("Baseline Training History")
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.savefig("importance_outputs/training_history_baseline.png", dpi=300)
plt.show()



# === Feature Importance Strategies ===
for strategy in strategies:
    print(f"\nEvaluating strategy: {strategy.upper()}")
    start_time = time.time()
    importances = get_feature_importance(strategy, model, X_test, y_test)
    duration = time.time() - start_time

    assert len(importances) == X_test.shape[1]
    assert not np.any(np.isnan(importances))
    assert not np.any(np.isinf(importances))

    strategy_importances[strategy] = importances

    plt.figure(figsize=(12, 4))
    sns.barplot(x=np.arange(len(importances)), y=importances)
    plt.title(f"Feature Importances – {strategy}")
    plt.tight_layout()
    plt.savefig(f"importance_outputs/importance_barplot_{strategy}.png", dpi=300)
    plt.close()

    important_idx = [i for i, imp in enumerate(importances) if imp > 0]
    selected_features = X.columns[important_idx]
    removed_features = X.columns.difference(selected_features)

    with open(f"importance_outputs/removed_features_{strategy}.txt", "w") as f:
        f.write("\n".join(removed_features))

    X_filt = scaler.fit_transform(X[selected_features])
    X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(X_filt, y, test_size=0.1, random_state=42)

    model_filt = build_model(X_train_f.shape[1])
    model_filt.fit(X_train_f, y_train_f, epochs=epochs, batch_size=32, validation_split=0.2, verbose=0)

    y_pred_f = (model_filt.predict(X_test_f) > 0.5).astype(int)
    y_proba_f = model_filt.predict(X_test_f)
    report_dict = classification_report(y_test_f, y_pred_f, digits=3, output_dict=True)
    classification_reports[strategy] = report_dict

    auc_score = roc_auc_score(y_test_f, y_proba_f)
    acc = model_filt.evaluate(X_test_f, y_test_f, verbose=0)[1]
    precision = precision_score(y_test_f, y_pred_f)
    recall = recall_score(y_test_f, y_pred_f)
    f1 = f1_score(y_test_f, y_pred_f)

    results.append({
        "Strategy": strategy,
        "Num_Features": len(selected_features),
        "Accuracy": round(acc, 3),
        "Precision": round(precision, 3),
        "Recall": round(recall, 3),
        "F1": round(f1, 3),
        "AUC": round(auc_score, 3),
        "Time (s)": round(duration, 2),
        "Removed_Features": len(removed_features)
    })

def plot_execution_time_and_removed_features(results_df):
    filtered = results_df[results_df['Strategy'] != 'baseline']
    x = np.arange(len(filtered))
    strategies = filtered['Strategy'].tolist()
    times = filtered['Time (s)'].tolist()
    removed = filtered['Removed_Features'].tolist()

    fig, ax1 = plt.subplots(figsize=(8, 6))
    ax2 = ax1.twinx()
    line1 = ax1.plot(x, times, color='#1f77b4', marker='o', label='Execution Time (s)')
    line2 = ax2.plot(x, removed, color='#C20078', marker='s', label='Removed Features')

    ax1.set_xlabel('Strategy')
    ax1.set_ylabel('Execution Time (s)', color='#1f77b4')
    ax2.set_ylabel('Number of Removed Features', color='#C20078')
    ax1.set_xticks(x)
    ax1.set_xticklabels(strategies, rotation=45)
    fig.suptitle('Execution Time and Removed Features')
    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    fig.legend(lines, labels, loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2)
    ax1.grid(axis='y', linestyle='--', alpha=0.6)
    ax2.grid(axis='y', linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.savefig("importance_outputs/execution_time_removed_features.png", dpi=300)
    plt.show()

# === Thresholded Strategy (Intersection) ===
all_importances = np.vstack([strategy_importances[s] for s in strategies])
min_mask = np.all(all_importances > threshold, axis=0)
final_selected_features = X.columns[min_mask]
print(f"\u2705 Selected {len(final_selected_features)} features using intersection strategy")

X_thresh = scaler.fit_transform(X[final_selected_features])
X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(X_thresh, y, test_size=0.1, random_state=42)
model_thresh = build_model(X_train_t.shape[1])
history_thresh = model_thresh.fit(X_train_t, y_train_t, epochs=epochs, batch_size=32, validation_split=0.2, verbose=1)

# === Thresholded Evaluation Report ===
y_pred_t = (model_thresh.predict(X_test_t) > 0.5).astype(int)
y_proba_t = model_thresh.predict(X_test_t)
report_thresh = classification_report(y_test_t, y_pred_t, digits=3, output_dict=True)
classification_reports['thresholded'] = report_thresh

auc_t = roc_auc_score(y_test_t, y_proba_t)
acc_t = model_thresh.evaluate(X_test_t, y_test_t, verbose=0)[1]
precision_t = precision_score(y_test_t, y_pred_t)
recall_t = recall_score(y_test_t, y_pred_t)
f1_t = f1_score(y_test_t, y_pred_t)

print("\n\U0001F4CA Thresholded Classification Report")
print(classification_report(y_test_t, y_pred_t, digits=3))

cm_thresh = confusion_matrix(y_test_t, y_pred_t)
disp_thresh = ConfusionMatrixDisplay(cm_thresh)
disp_thresh.plot(cmap="Greens", values_format=".0f")
plt.title("Thresholded Confusion Matrix")
plt.tight_layout()
plt.savefig("importance_outputs/confusion_matrix_thresholded.png", dpi=300)
plt.show()

plt.figure()
plt.plot(history_thresh.history['accuracy'], label='Train Accuracy')
plt.plot(history_thresh.history['val_accuracy'], label='Validation Accuracy')
plt.plot(history_thresh.history['loss'], label='Train Loss')
plt.plot(history_thresh.history['val_loss'], label='Validation Loss')
plt.title("Thresholded Training History")
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.savefig("importance_outputs/training_history_thresholded.png", dpi=300)
plt.show()

# === SHAP Visualization ===
background_df = pd.DataFrame(X_train_t, columns=final_selected_features)
X_sample_df = pd.DataFrame(X_test_t[:100], columns=final_selected_features)
background_df_sampled = shap.sample(background_df, 100, random_state=42)

explainer = shap.KernelExplainer(lambda x: model_thresh.predict(x).flatten(), background_df_sampled)
raw_shap_values = explainer.shap_values(X_sample_df)

shap_values = shap.Explanation(
    values=raw_shap_values,
    base_values=np.array(explainer.expected_value).flatten(),
    data=X_sample_df,
    feature_names=list(final_selected_features)
)

plt.figure(figsize=(15, 10))
shap.plots.beeswarm(shap_values, max_display=len(final_selected_features))
plt.savefig("importance_outputs/beeswarm_plot.png", dpi=300, bbox_inches='tight')
plt.show()

plt.figure()
shap.plots.bar(shap_values, max_display=len(final_selected_features))
plt.savefig("importance_outputs/bar_plot.png", dpi=300, bbox_inches='tight')
plt.show()

mean_abs_shap = np.abs(shap_values.values).mean(axis=0)
sorted_indices = np.argsort(mean_abs_shap)[::-1]
shap_df = pd.DataFrame({
    "Feature": np.array(final_selected_features)[sorted_indices],
    "Mean_Abs_SHAP": mean_abs_shap[sorted_indices]
})
shap_df.to_csv("importance_outputs/sorted_shap_features.csv", index=False)

# === Save Summary Table ===
results_df = pd.DataFrame(results)
results_df.to_csv("importance_outputs/feature_importance_comparison.csv", index=False)

plot_execution_time_and_removed_features(results_df)
```

```python
#Keep _modma
import os
import time
import shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (classification_report, roc_auc_score, confusion_matrix, 
                             precision_score, recall_score, f1_score, ConfusionMatrixDisplay, roc_curve)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from feature_importance_keras import get_feature_importance

# === Setup ===
os.makedirs("importance_outputs", exist_ok=True)
sns.set(style="whitegrid")
epochs = 30
strategies = ['3m', 'permutation', 'median', 'mean', 'max', 'zero','hybrid']
threshold = 0.0
classification_reports = {}
results = []
strategy_importances = {}

# === Load Data ===
data = pd.read_csv('Data_modma/all_audio_frames_features.csv')
X = data.iloc[:, :-2]
y = data.iloc[:, -1]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.1, random_state=42)

# === Model Builder ===
def build_model(input_dim):
    model = Sequential([
        Dense(128, input_dim=input_dim, activation='relu'),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# === Baseline Model ===
print("Training baseline model...")
model = build_model(X_train.shape[1])
history_base = model.fit(X_train, y_train, epochs=epochs, batch_size=32, validation_split=0.2, verbose=1)

y_pred = (model.predict(X_test) > 0.5).astype(int)
y_proba = model.predict(X_test)
initial_report = classification_report(y_test, y_pred, digits=3, output_dict=True)
classification_reports['baseline'] = initial_report

initial_auc = roc_auc_score(y_test, y_proba)
initial_acc = model.evaluate(X_test, y_test, verbose=0)[1]
initial_precision = precision_score(y_test, y_pred)
initial_recall = recall_score(y_test, y_pred)
initial_f1 = f1_score(y_test, y_pred)

# === Baseline Evaluation Report ===
print("\n\U0001F4CA Baseline Classification Report")
print(classification_report(y_test, y_pred, digits=3))

cm_base = confusion_matrix(y_test, y_pred)
disp_base = ConfusionMatrixDisplay(cm_base)
disp_base.plot(cmap="Blues", values_format=".0f")
plt.title("Baseline Confusion Matrix")
plt.tight_layout()
plt.savefig("importance_outputs/confusion_matrix_baseline.png", dpi=300)
plt.show()

plt.figure()
plt.plot(history_base.history['accuracy'], label='Train Accuracy')
plt.plot(history_base.history['val_accuracy'], label='Validation Accuracy')
plt.plot(history_base.history['loss'], label='Train Loss')
plt.plot(history_base.history['val_loss'], label='Validation Loss')
plt.title("Baseline Training History")
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.savefig("importance_outputs/training_history_baseline.png", dpi=300)
plt.show()



# === Feature Importance Strategies ===
for strategy in strategies:
    print(f"\nEvaluating strategy: {strategy.upper()}")
    start_time = time.time()
    importances = get_feature_importance(strategy, model, X_test, y_test)
    duration = time.time() - start_time

    assert len(importances) == X_test.shape[1]
    assert not np.any(np.isnan(importances))
    assert not np.any(np.isinf(importances))

    strategy_importances[strategy] = importances

    plt.figure(figsize=(12, 4))
    sns.barplot(x=np.arange(len(importances)), y=importances)
    plt.title(f"Feature Importances – {strategy}")
    plt.tight_layout()
    plt.savefig(f"importance_outputs/importance_barplot_{strategy}.png", dpi=300)
    plt.close()

    important_idx = [i for i, imp in enumerate(importances) if imp > 0]
    selected_features = X.columns[important_idx]
    removed_features = X.columns.difference(selected_features)

    with open(f"importance_outputs/removed_features_{strategy}.txt", "w") as f:
        f.write("\n".join(removed_features))

    X_filt = scaler.fit_transform(X[selected_features])
    X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(X_filt, y, test_size=0.1, random_state=42)

    model_filt = build_model(X_train_f.shape[1])
    model_filt.fit(X_train_f, y_train_f, epochs=epochs, batch_size=32, validation_split=0.2, verbose=0)

    y_pred_f = (model_filt.predict(X_test_f) > 0.5).astype(int)
    y_proba_f = model_filt.predict(X_test_f)
    report_dict = classification_report(y_test_f, y_pred_f, digits=3, output_dict=True)
    classification_reports[strategy] = report_dict

    auc_score = roc_auc_score(y_test_f, y_proba_f)
    acc = model_filt.evaluate(X_test_f, y_test_f, verbose=0)[1]
    precision = precision_score(y_test_f, y_pred_f)
    recall = recall_score(y_test_f, y_pred_f)
    f1 = f1_score(y_test_f, y_pred_f)

    results.append({
        "Strategy": strategy,
        "Num_Features": len(selected_features),
        "Accuracy": round(acc, 3),
        "Precision": round(precision, 3),
        "Recall": round(recall, 3),
        "F1": round(f1, 3),
        "AUC": round(auc_score, 3),
        "Time (s)": round(duration, 2),
        "Removed_Features": len(removed_features)
    })

def plot_execution_time_and_removed_features(results_df):
    filtered = results_df[results_df['Strategy'] != 'baseline']
    x = np.arange(len(filtered))
    strategies = filtered['Strategy'].tolist()
    times = filtered['Time (s)'].tolist()
    removed = filtered['Removed_Features'].tolist()

    fig, ax1 = plt.subplots(figsize=(8, 6))
    ax2 = ax1.twinx()
    line1 = ax1.plot(x, times, color='#1f77b4', marker='o', label='Execution Time (s)')
    line2 = ax2.plot(x, removed, color='#C20078', marker='s', label='Removed Features')

    ax1.set_xlabel('Strategy')
    ax1.set_ylabel('Execution Time (s)', color='#1f77b4')
    ax2.set_ylabel('Number of Removed Features', color='#C20078')
    ax1.set_xticks(x)
    ax1.set_xticklabels(strategies, rotation=45)
    fig.suptitle('Execution Time and Removed Features')
    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    fig.legend(lines, labels, loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2)
    ax1.grid(axis='y', linestyle='--', alpha=0.6)
    ax2.grid(axis='y', linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.savefig("importance_outputs/execution_time_removed_features.png", dpi=300)
    plt.show()

# === Thresholded Strategy (Intersection) ===
all_importances = np.vstack([strategy_importances[s] for s in strategies])
min_mask = np.all(all_importances > threshold, axis=0)
final_selected_features = X.columns[min_mask]
print(f"\u2705 Selected {len(final_selected_features)} features using intersection strategy")

X_thresh = scaler.fit_transform(X[final_selected_features])
X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(X_thresh, y, test_size=0.1, random_state=42)
model_thresh = build_model(X_train_t.shape[1])
history_thresh = model_thresh.fit(X_train_t, y_train_t, epochs=epochs, batch_size=32, validation_split=0.2, verbose=1)

# === Thresholded Evaluation Report ===
y_pred_t = (model_thresh.predict(X_test_t) > 0.5).astype(int)
y_proba_t = model_thresh.predict(X_test_t)
report_thresh = classification_report(y_test_t, y_pred_t, digits=3, output_dict=True)
classification_reports['thresholded'] = report_thresh

auc_t = roc_auc_score(y_test_t, y_proba_t)
acc_t = model_thresh.evaluate(X_test_t, y_test_t, verbose=0)[1]
precision_t = precision_score(y_test_t, y_pred_t)
recall_t = recall_score(y_test_t, y_pred_t)
f1_t = f1_score(y_test_t, y_pred_t)

print("\n\U0001F4CA Thresholded Classification Report")
print(classification_report(y_test_t, y_pred_t, digits=3))

cm_thresh = confusion_matrix(y_test_t, y_pred_t)
disp_thresh = ConfusionMatrixDisplay(cm_thresh)
disp_thresh.plot(cmap="Greens", values_format=".0f")
plt.title("Thresholded Confusion Matrix")
plt.tight_layout()
plt.savefig("importance_outputs/confusion_matrix_thresholded.png", dpi=300)
plt.show()

plt.figure()
plt.plot(history_thresh.history['accuracy'], label='Train Accuracy')
plt.plot(history_thresh.history['val_accuracy'], label='Validation Accuracy')
plt.plot(history_thresh.history['loss'], label='Train Loss')
plt.plot(history_thresh.history['val_loss'], label='Validation Loss')
plt.title("Thresholded Training History")
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.savefig("importance_outputs/training_history_thresholded.png", dpi=300)
plt.show()

# === SHAP Visualization ===
background_df = pd.DataFrame(X_train_t, columns=final_selected_features)
X_sample_df = pd.DataFrame(X_test_t[:100], columns=final_selected_features)
background_df_sampled = shap.sample(background_df, 100, random_state=42)

explainer = shap.KernelExplainer(lambda x: model_thresh.predict(x).flatten(), background_df_sampled)
raw_shap_values = explainer.shap_values(X_sample_df)

shap_values = shap.Explanation(
    values=raw_shap_values,
    base_values=np.array(explainer.expected_value).flatten(),
    data=X_sample_df,
    feature_names=list(final_selected_features)
)

plt.figure(figsize=(15, 10))
shap.plots.beeswarm(shap_values, max_display=len(final_selected_features))
plt.savefig("importance_outputs/beeswarm_plot.png", dpi=300, bbox_inches='tight')
plt.show()

plt.figure()
shap.plots.bar(shap_values, max_display=len(final_selected_features))
plt.savefig("importance_outputs/bar_plot.png", dpi=300, bbox_inches='tight')
plt.show()

mean_abs_shap = np.abs(shap_values.values).mean(axis=0)
sorted_indices = np.argsort(mean_abs_shap)[::-1]
shap_df = pd.DataFrame({
    "Feature": np.array(final_selected_features)[sorted_indices],
    "Mean_Abs_SHAP": mean_abs_shap[sorted_indices]
})
shap_df.to_csv("importance_outputs/sorted_shap_features.csv", index=False)

# === Save Summary Table ===
results_df = pd.DataFrame(results)
results_df.to_csv("importance_outputs/feature_importance_comparison.csv", index=False)

plot_execution_time_and_removed_features(results_df)
```

```python
import os
import time
import shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (classification_report, roc_auc_score, confusion_matrix, 
                             precision_score, recall_score, f1_score, ConfusionMatrixDisplay, roc_curve)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from feature_importance_keras import get_feature_importance

# === 56-feature Manual Full Set ===
selected_features_full = [
    'mfcc_mean_0', 'spectral_mean_2', 'mfcc_mean_9', 'spectral_mean_4', 'mfcc_mean_11',
    'f0_mean', 'spectral_mean_1', 'mfcc_mean_6', 'mfcc_mean_8', 'spectral_mean_3',
    'mfcc_mean_2', 'mfcc_mean_7', 'mfcc_mean_5', 'spectral_mean_5', 'mfcc_std_2',
    'mfcc_mean_4', 'mfcc_mean_3', 'spectral_mean_6', 'spectral_std_5', 'mfcc_std_3',
    'mfcc_mean_1', 'rms_std', 'spectral_std_6', 'spectral_mean_0', 'chroma_mean_3',
    'mfcc_std_0', 'zero_crossing_rate', 'chroma_mean_7', 'mfcc_std_9', 'mfcc_std_4',
    'mfcc_std_6', 'mfcc_std_5', 'mfcc_std_1', 'spectral_std_0', 'mfcc_delta2_std_7',
    'chroma_mean_5', 'mfcc_delta2_std_3', 'mfcc_std_8', 'mfcc_delta_std_11',
    'chroma_mean_9', 'chroma_std_11', 'mfcc_delta_std_2', 'chroma_std_1',
    'mfcc_delta2_std_11', 'chroma_std_9', 'mfcc_delta_std_0', 'mfcc_delta_mean_4',
    'chroma_std_10', 'chroma_std_4', 'mfcc_delta2_std_6', 'mfcc_delta2_mean_7',
    'chroma_std_0', 'mfcc_delta2_mean_1', 'mfcc_delta_mean_6', 'mfcc_delta2_std_8',
    'mfcc_delta2_mean_8'
]

# === Setup ===
os.makedirs("importance_outputs", exist_ok=True)
sns.set(style="whitegrid")
epochs = 50
strategies = ['3m', 'permutation', 'median', 'mean', 'max', 'zero', 'hybrid']
threshold = 0.0
ablation_results = []
n_repeats = 5

# === Load Data ===
data = pd.read_csv('Data_edaic/04_sec.csv')
X = data.iloc[:, :-2]
y = data.iloc[:, -1]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.1, random_state=42)

# === Model Builder ===
def build_model(input_dim):
    model = Sequential([
        Dense(128, input_dim=input_dim, activation='relu'),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# === Train Full Model Once (For Importance Computation) ===
model = build_model(X_train.shape[1])
model.fit(X_train, y_train, epochs=epochs, batch_size=32, validation_split=0.2, verbose=0)

# === Compute Feature Importances ===
strategy_importances = {}
for strategy in strategies:
    print(f"\nEvaluating strategy: {strategy.upper()}")
    importances = get_feature_importance(strategy, model, X_test, y_test)
    strategy_importances[strategy] = importances

# === Ablation Function with Repeats ===
def run_ablation(strategy_subset, label, override_features=None):
    print(f"\nRunning ablation: {label}")
    accs, precs, recs, f1s, aucs, times, params = [], [], [], [], [], [], []
    num_feats = 0

    for i in range(n_repeats):
        if override_features is not None:
            selected = override_features
        else:
            imp_subset = np.vstack([strategy_importances[s] for s in strategy_subset])
            mask = np.all(imp_subset > threshold, axis=0)
            selected = X.columns[mask]

        if len(selected) == 0:
            print(f"[{label}] No features selected. Skipping.")
            return

        X_sel = scaler.fit_transform(X[selected])
        X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(X_sel, y, test_size=0.1, random_state=42 + i)

        model_a = build_model(X_train_a.shape[1])
        start_time = time.time()
        model_a.fit(X_train_a, y_train_a, epochs=epochs, batch_size=32, validation_split=0.2, verbose=0)
        elapsed_time = time.time() - start_time

        y_pred_a = (model_a.predict(X_test_a) > 0.5).astype(int)
        y_proba_a = model_a.predict(X_test_a)

        auc_a = roc_auc_score(y_test_a, y_proba_a)
        acc_a = model_a.evaluate(X_test_a, y_test_a, verbose=0)[1]
        precision_a = precision_score(y_test_a, y_pred_a)
        recall_a = recall_score(y_test_a, y_pred_a)
        f1_a = f1_score(y_test_a, y_pred_a)

        accs.append(acc_a)
        precs.append(precision_a)
        recs.append(recall_a)
        f1s.append(f1_a)
        aucs.append(auc_a)
        times.append(elapsed_time)
        params.append(model_a.count_params())
        num_feats = len(selected)

    # Rename label for display
    if label == "Full":
        label = "ALL"
    elif label == "No-Selection":
        label = "None"

    ablation_results.append({
        "Condition": label,
        "Strategies_Used": "+".join(strategy_subset) if override_features is None else "manual",
        "Num_Features": num_feats,
        "Removed_Features": X.shape[1] - num_feats,
        "Accuracy (mean±std)": f"{np.mean(accs):.3f} ± {np.std(accs):.3f}",
        "Precision (mean±std)": f"{np.mean(precs):.3f} ± {np.std(precs):.3f}",
        "Recall (mean±std)": f"{np.mean(recs):.3f} ± {np.std(recs):.3f}",
        "F1 (mean±std)": f"{np.mean(f1s):.3f} ± {np.std(f1s):.3f}",
        "AUC (mean±std)": f"{np.mean(aucs):.3f} ± {np.std(aucs):.3f}",
        "Model Size (params)": int(np.mean(params)),
        "Time (s)": f"{np.mean(times):.2f}"
    })

# === Run Only Desired Ablation Conditions ===

# Manual full set → renamed as ALL
run_ablation(strategies, "Full", override_features=selected_features_full)

# Single-strategy only
for strategy in strategies:
    run_ablation([strategy], f"Only-{strategy}")

# No Feature Selection → renamed as None
run_ablation([], "No-Selection", override_features=list(X.columns))

# === Save and Display Results ===
ablation_df = pd.DataFrame(ablation_results)
ablation_df.to_csv("importance_outputs/ablation_study_results.csv", index=False)

from IPython.display import display
display(ablation_df)
```
